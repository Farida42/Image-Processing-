```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt

# Display original image
image = Image.open("/content/pizza.png")
plt.figure(figsize=(8, 8))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image)
plt.axis('off')

# Enhance brightness
enhancer = ImageEnhance.Brightness(image)
enhanced_img = enhancer.enhance(1.5)
plt.subplot(1, 3, 2)
plt.title("Increased Brightness")
plt.imshow(enhanced_img)
plt.axis('off')

# Decrease brightness
factor = 0.3  # Decrease brightness
dark_img = enhancer.enhance(factor)
plt.subplot(1, 3, 3)
plt.title("Decreased Brightness")
plt.imshow(dark_img)
plt.axis('off')

plt.show()
```

```python
from PIL import Image

# Open the image
image = Image.open("/content/pizza.png")

# Adding a constant (increasing brightness)
k_add = 50
added_image = Image.eval(image, lambda x: x + k_add)

# Subtracting a constant (decreasing brightness)
k_subtract = 50
subtracted_image = Image.eval(image, lambda x: x - k_subtract)

# Negative of the image
negative_image = Image.eval(image, lambda x: 255 - x)

# Multiply by constant
k_multiply = 1.5
multiplied_image = Image.eval(image, lambda x: x * k_multiply)

# Divide by constant
k_divide = 2
divided_image = Image.eval(image, lambda x: x / k_divide)

# Divide into constant
k_divide_into = 200
divided_into_image = Image.eval(image, lambda x: k_divide_into / x if x != 0 else 255)

# OR constant
k_or = 100
or_image = Image.eval(image, lambda x: x | k_or)

# AND constant
k_and = 150
and_image = Image.eval(image, lambda x: x & k_and)

# XOR constant
k_xor = 200
xor_image = Image.eval(image, lambda x: x ^ k_xor)

# Display the images
added_image.show()
subtracted_image.show()
negative_image.show()
multiplied_image.show()
divided_image.show()
divided_into_image.show()
or_image.show()
and_image.show()
xor_image.show()
```
