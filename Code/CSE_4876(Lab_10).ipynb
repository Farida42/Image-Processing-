```python
import heapq
from collections import defaultdict
import numpy as np
import cv2

# Step 3: Calculate Pixel Frequencies
def calculate_frequencies(image):
    frequencies = defaultdict(int)
    for row in image:
        for pixel in row:
            frequencies[pixel] += 1
    return frequencies

# Step 4: Build Huffman Tree
def build_huffman_tree(frequencies):
    heap = [[weight, [pixel, ""]] for pixel, weight in frequencies.items()]
    heapq.heapify(heap)
    while len(heap) > 1:
        lo = heapq.heappop(heap)
        hi = heapq.heappop(heap)
        for pair in lo[1:]:
            pair[1] = '0' + pair[1]
        for pair in hi[1:]:
            pair[1] = '1' + pair[1]
        heapq.heappush(heap, [lo[0] + hi[0]] + lo[1:] + hi[1:])
    return heap[0]

# Step 5: Generate Huffman Codes
def generate_huffman_codes(root):
    codes = {}
    stack = root[1:]
    while stack:
        node = stack.pop()
        if isinstance(node[0], tuple):
            stack.extend(node[1:])
        else:
            codes[node[0]] = node[1]
    return codes

# Step 6: Encode Image
def encode_image(image, codes):
    encoded_image = ""
    for row in image:
        for pixel in row:
            encoded_image += codes[pixel]
    return encoded_image

# Step 7: Write Compressed Image to File
def write_compressed_image(encoded_image, output_file):
    with open(output_file, 'w') as f:
        f.write(encoded_image)

# Step 8: Decode Image (Optional)
def decode_image(encoded_image, codes):
    decoded_image = ""
    code = ""
    for bit in encoded_image:
        code += bit
        if code in codes.values():
            pixel = [key for key, value in codes.items() if value == code][0]
            decoded_image += str(pixel)
            code = ""
    return decoded_image

# Read Image
image = cv2.imread("input_image.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 2: Convert Image to Pixel Array
pixel_array = image.reshape(-1, 3)

# Step 3: Calculate Pixel Frequencies
frequencies = calculate_frequencies(pixel_array)

# Step 4: Build Huffman Tree
huffman_tree = build_huffman_tree(frequencies)

# Step 5: Generate Huffman Codes
codes = generate_huffman_codes(huffman_tree)

# Step 6: Encode Image
encoded_image = encode_image(pixel_array, codes)

# Step 7: Write Compressed Image to File
write_compressed_image(encoded_image, "compressed_image.txt")

# Step 8: Decode Image (Optional)
decoded_image = decode_image(encoded_image, {v: k for k, v in codes.items()})
decoded_image = np.array(list(decoded_image)).reshape(image.shape[0], image.shape[1], 3).astype(np.uint8)

# Save decoded image
cv2.imwrite("decoded_image.jpg", decoded_image)
```
