# **Temporal processing**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img1 = cv2.imread("/content/c.png")
img2 = cv2.imread("/content/c1.png")

cv2_imshow(img1)
cv2_imshow(img2)
```

```python
img1 = cv2.imread("/content/c.png")
img2 = cv2.imread("/content/c1.png")
# Display images side by side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Image 2')
plt.axis('off')

plt.show()
```

```python
img1 = cv2.resize(img1, (300,300))
img2 = cv2.resize(img2, (300,300))

cv2_imshow(img1)
cv2_imshow(img2)
```

```python
# Resize images to 300x300
img1 = cv2.resize(img1, (300, 300))
img2 = cv2.resize(img2, (300, 300))

# Display images side by side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Image 1 (Resized)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Image 2 (Resized)')
plt.axis('off')

plt.show()
```

**Adding**

```python
img_add = cv2.add(img1,img2)
cv2_imshow(img_add)
```

**Subtracting**

```python
# Subtracting the images
img_subtract = cv2.subtract(img1, img2)

# Display the subtracted image
cv2_imshow(img_subtract)
```

**Multiplying**

```python
# Multiplying the images
img_multiply = cv2.multiply(img1, img2)

# Display the multiplied image
cv2_imshow(img_multiply)
```

**Signal Averaging**

```python
# Compute signal averaging
img_avg = cv2.addWeighted(img1, 0.6, img2, 0.6, 0)

# Display the averaged image
cv2_imshow(img_avg)
```

**AND**

```python
# Perform logical AND operation
img_and = cv2.bitwise_and(img1, img2)

# Display the ANDed image
cv2_imshow(img_and)
```

**OR**

```python
# Perform logical OR operation
img_or = cv2.bitwise_or(img1, img2)

# Display the ANDed image
cv2_imshow(img_or)
```

**XOR**

```python
# Perform logical OR operation
img_Xor = cv2.bitwise_xor(img1, img2)

# Display the ANDed image
cv2_imshow(img_Xor)
```

# **Geometric Processing**

## **Clipping**

```python
# Read the image
img = cv2.imread('/content/c.png')

# Clipping
clipped_img = np.clip(img, 100, 200)  # Adjust the range as needed

# Display results
plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(clipped_img, cv2.COLOR_BGR2RGB))
plt.title('Clipped Image')
plt.axis('off')
```

## **Rotating**

```python
# Rotating
rows, cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)  # Rotate 45 degrees
rotated_img = cv2.warpAffine(img, M, (cols, rows))

# Display results
plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')
plt.axis('off')
```

# **1.Clipping 2.Rotating**

```python
# Read the image
img = cv2.imread('/content/c.png')

# Clipping
clipped_img = np.clip(img, 100, 200)  # Adjust the range as needed

# Rotating
rows, cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)  # Rotate 45 degrees
rotated_img = cv2.warpAffine(img, M, (cols, rows))

# Display results
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(clipped_img, cv2.COLOR_BGR2RGB))
plt.title('Clipped Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')
plt.axis('off')

plt.show()
```
